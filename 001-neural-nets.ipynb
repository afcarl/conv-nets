A single-layered neural net is simply a set of weights 

```python
fig, ax = plt.subplots(1)
layer_sizes = [10,5, 2]
left=.1
right=.9
bottom=.1
top=.9

v_spacing = (top - bottom)/float(max(layer_sizes))
h_spacing = (right - left)/float(len(layer_sizes) - 1)
# Nodes
for n, layer_size in enumerate(layer_sizes):
    layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
    for m in range(layer_size):
        circle = plt.Circle((n*h_spacing + left, layer_top - m*v_spacing),
                            v_spacing / 4.,
                            color='w', ec='k', zorder=4)
        ax.add_artist(circle)
        t = ax.text(circle.center[0]-0.02, circle.center[1], "$X_{%s%s}$"%(n+1, m+1))
        t.set_zorder(10)
# Edges
for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1],
                                                 layer_sizes[1:])):
    layer_top_a = v_spacing*(layer_size_a - 1) / 2. + (top + bottom) / 2.
    layer_top_b = v_spacing*(layer_size_b - 1) / 2. + (top + bottom) / 2.
    for m in range(layer_size_a):
        for o in range(layer_size_b):
            line = plt.Line2D([n*h_spacing + left, (n + 1) * h_spacing +
                               left],
                              [layer_top_a - m*v_spacing, layer_top_b -
                               o*v_spacing], c='k')
            ax.add_artist(line)
```

A multi-layered network will add to that another set of weights: 

```python
weights_input = [[1, 2], [3, 4]]
weights_hidden = [3, 4]
```

```python
hidden1 = np.dot(weights_input[0], data)
hidden2 = np.dot(weights_input[1], data)
```

```python
output = np.dot(weights_hidden, [hidden1, hidden2])
```

So far, this is just a composition of linear transformations! So, it could be written as:
